# Run PC Analysis on all datasets generated

```python
# define top-level directory for datasets
data_dir = 'simulated_data'

# Find all dataset files
all_data_files = sorted(glob.glob(f"{data_dir}/*/*_data.csv"))
# Find all graph.npy files and sort them
all_graph_files = sorted(glob.glob(f"{data_dir}/*/*_graph.npy"))

# check if found same number of data and graph files
if len(all_data_files) == 0:
    print(f"Error: No datasets found in directory '{data_dir}'.")
    print("Please check the 'data_dir' variable.")
elif len(all_data_files) != len(all_graph_files):
    print("Error: # of data files and graph files does not match")
else:
    print(f"Found {len(all_data_files)} datasets. Starting analysis:")
    
# Store results in a dictionary
results_summary = {}

    # Loop through all files and run the analysis
    # using zip() to pair each data file with its corresponding graph file
for data_file, graph_file in zip(all_data_files, all_graph_files):
        
    result = run_pc_analysis_on_dataset(data_file, graph_file, alpha=0.05)
        
        # store SHD score
    dataset_name = os.path.basename(data_file).replace('_data.csv', '')
    results_summary[dataset_name] = result['shd_score']

    #  Print a final summary of all scores 
print("FINAL RESULTS SUMMARY (SHD)")

for name, score in results_summary.items():
    print(f"{name}: {score}")
```
